```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.ticker as ticker
```

```python
tabla1 = pd.read_csv("https://raw.githubusercontent.com/DrRobert12/Challenge-Alura-Robert/refs/heads/main/1%C2%B0%20Challenge%20ALURA%20DATA%20SCIENCE/tienda1.csv")
tabla2 = pd.read_csv("https://raw.githubusercontent.com/DrRobert12/Challenge-Alura-Robert/refs/heads/main/1%C2%B0%20Challenge%20ALURA%20DATA%20SCIENCE/tienda2.csv")
tabla3 = pd.read_csv("https://raw.githubusercontent.com/DrRobert12/Challenge-Alura-Robert/refs/heads/main/1%C2%B0%20Challenge%20ALURA%20DATA%20SCIENCE/tienda3.csv")
tabla4 = pd.read_csv("https://raw.githubusercontent.com/DrRobert12/Challenge-Alura-Robert/refs/heads/main/1%C2%B0%20Challenge%20ALURA%20DATA%20SCIENCE/tienda4.csv")
```

### FACTURACION TIENDA 1

```python
# Agrupar y ordenar por suma de precios de mayor a menor
categoria_precio = tabla1.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)

# Gráfico
plt.figure(figsize=(9, 7)) # El primer numero es el ancho y el segundo la altura
bars = plt.bar(categoria_precio.index, categoria_precio.values, color='green', alpha=0.7, edgecolor='red')

# Etiquetas de cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:,.0f}', ha='center', va='bottom', fontsize=9)

plt.title('Facturacion de Productos por Categoría')
plt.xlabel('Categoría del Producto')
plt.ylabel('Precios')
plt.xticks(rotation=80)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
# print Gráfico
plt.show()


# # Graficar la facturacion total de la tienda 1
# plt.figure(figsize=(6, 4)) # El primer numero es el ancho y el segundo la altura

# bars = plt.bar(['Tienda 1'], [precio_total], color='green', alpha=0.7, edgecolor='red')

# for bar in bars:
#     yval = bar.get_height()
#     plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:,.0f}', ha='center', va='bottom', fontsize=9)
# plt.title('Facturacion Total de la Tienda 1')
# plt.xticks(rotation=360)
# plt.grid(axis='y', linestyle='--', alpha=0.7)
# plt.tight_layout()
# plt.show()





```

### CATEGORÍA MÁS POPULARES

```python
# Concatenar las 4 tablas en una sola
tabla_general = pd.concat([tabla1, tabla2, tabla3, tabla4], ignore_index=True)
categoria_popu = tabla_general.groupby('Categoría del Producto')['Calificación'].sum().sort_values(ascending=False)

# Gráfico
plt.figure(figsize=(9, 7)) # El primer numero es el ancho y el segundo la altura
bars = plt.bar(categoria_popu.index, categoria_popu.values, color='gold', alpha=0.7, edgecolor='indigo')

# Etiquetas de cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:,.0f}', ha='center', va='bottom', fontsize=9)

plt.title('Categorías más Populares')
plt.xlabel('Categoría del Producto')
plt.ylabel('Suma de Calificaciones')
plt.xticks(rotation=80)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# print Gráfico 
plt.show()
```

### PROMEDIO CALIFICACIÓN DE PRODUCTOS

```python
promedio_productos = tabla_general.groupby('Producto')['Calificación'].mean().sort_values(ascending=False)

# Gráfico
plt.figure(figsize=(12, 5)) # El primer numero es el ancho y el segundo la altura
bars = plt.bar(promedio_productos.index, promedio_productos.values, color='coral', alpha=0.7, edgecolor='violet')

# Etiquetas de cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:,.2f}', ha='center', va='bottom', fontsize=6)
plt.subplots_adjust(bottom=0.7)  # Ajustar el espacio inferior para evitar que las etiquetas se corten
plt.title('Promedio de Calificaciones')
plt.xlabel('Producto')
plt.ylabel('Promedio')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# print Gráfico 
plt.show()

```

### PROMEDIO CALIFICACIÓN DE CLIENTES

```python
promedio_clientes = tabla_general.groupby('Vendedor')['Calificación'].mean().sort_values(ascending=False)

# Gráfico
plt.figure(figsize=(12, 5)) # El primer numero es el ancho y el segundo la altura
bars = plt.bar(promedio_clientes.index, promedio_clientes.values, color='lime', alpha=0.7, edgecolor='tan')

# Etiquetas de cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:,.2f}', ha='center', va='bottom', fontsize=8)
plt.subplots_adjust(bottom=0.7)  # Ajustar el espacio inferior para evitar que las etiquetas se corten
plt.title('Promedio de Calificaciones de Clientes')
plt.xlabel('Vendedor')
plt.ylabel('Promedio')
plt.xticks(rotation=70)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# print Gráfico 
plt.show()
```

### PRODUCTOS MAS Y MENOS VENDIDOS
